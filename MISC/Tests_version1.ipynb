```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
graph =  nx.Graph()
graph.add_nodes_from(["john","alex","mambo","cheche","newton","stevo","kennedy"])
import matplotlib.pyplot as plt
nx.draw(graph, with_labels = True)
plt.show()
```

```python
graph.add_edges_from([
    ("john","alex"),("john","kennedy"),("john","mambo"),("john","cheche"),("john","stevo"),
    ("alex","kennedy"),("alex","kennedy"),("john","cheche"),("john","newton"),
    ("mambo","cheche"),("mambo","alex"),("cheche","alex"),
    ("newton","kennedy"),("newton","stevo")
])
```

```python
nx.draw(graph, with_labels = True)
plt.show()
```

```python
# calculate the degree of centarity
central = nx.degree_centrality(graph)
dt = pd.DataFrame(central.keys(),columns = ["nodes"])
dt["degree"] = central.values()
dt.sort_values("degree",ascending = False)
```
