```python
import os
import pandas as pd
os.listdir('preliminary data')
df = pd.read_excel('preliminary data/final_fiji.xlsx')
```

```python
import networkx as nx
df.head()
```

```python
firms = list(df.firm_id.unique())[:10]
G = nx.Graph()
nodes = list(df.firm_id.unique())+list(df.answer.unique())
G.add_nodes_from(nodes)
G.nodes()
```

```python
edges = []
for i,j,k in zip(df["firm_id"],df["answer"],df['survey']):
    edges.append((i,j,{'survey':k}))
G.add_edges_from(edges)
G.edges(data = True)
```

```python
for i in G.nodes():
    if i in df["firm_id"].unique():
        G.nodes()[i]["color"] = "#00AE9F"
    else:
        G.nodes()[i]["color"] = "#EA5B0C"
G.nodes(data = True)
```

```python
nx.draw(G,
        node_color = [i[1]["color"] for i in G.nodes(data = True)])
```

```python
nx.write_graphml(G,"fijiV1.graphml")
```
